```python
def sub_mean(df):
    df_means = df.mean(axis = 1, skipna = True)
    
    for i in range(len(df)):
        df.iloc[[i]] = df.iloc[[i]]-df_means.iloc[[i]]
        
    return df
```

```python
import numpy as np
from numpy.linalg import norm

def cosine_sim(df, user):
    cosine_sim_val = np.zeros(len(df))
    for i in range(len(df)):
        cosine_sim_val[i] = np.dot(np.array(df.iloc[[user]]),np.array(df.iloc[[i]]))/(norm(np.array(df.iloc[[user]]))*norm(np.array(df.iloc[[i]])))
    df['cosine_sim'] = cosine_sim_val
    return df
    
```
